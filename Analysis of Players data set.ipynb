```python
# Qs 5. Convert the following 
      # 1.height of the players in cms  
    
     
```

```python
# Intialise a list as Height_cm 
Height_cm = [] 

# Iterate to height string value in the data frame
for i in list(df['Height'].values):
    try:
        # height is in string and we are converting it into numerical values by string conversion to float.
        Height_cm.append((float(str(i)[0])*12.0 + float(str(i)[2:]))*2.54) 
    except(ValueError): 
        # Their can be values in the string height which are non numeric or missing values in string 
        # whenever a nul value is recieved it will be appended using np.nan (Not a number)
        Height_cm.append(np.nan) 
        
df['Height_cm'] = Height_cm

```

```python
 # 2.Weight of players in Kg
    # string typecast to float and convert weight into kg by dividing from 2.204
df['Weight_kg'] = df['Weight'].str[:3].astype(float)/2.204 
print(df.head())
```

```python
import seaborn as sns    
import matplotlib.pyplot as plt
# create a function get_statistic which will calculate mean, standard deviation for any user defined attribute
def get_statistics(numeric_column_name): 
    print("Mean {}: ".format(numeric_column_name),   
          df[numeric_column_name].mean())
    print("Standard Deviation in {}: ".format(numeric_column_name), 
          df[numeric_column_name].std())
    sns.set()
    plt.title("{} Histogram".format(numeric_column_name))
    df[numeric_column_name].hist(bins = 10)
    
# Function Height_cm is called    
get_statistics('Height_cm')
   
```

```python
# part C) Data Manipulation .
# Qs7(a) Which are the top 5 countries which send the highest number of players. 
# Qs7(b)Which are the top 5 positions.

# Part D) Data Visualization 
# Qs 8 Import Counter from collection library and create a user generated function to plot the top 5 positions. 

```

```python
# We can look at the frequency in Nationality of each player using the ‘Counter’ method
# A Counter is a container that keeps track of how many times equivalent values are added.
# counter is a special type of object data-set provided with the collections in python, 
from collections import Counter 
print(Counter(df['Nationality'].values)) 
```

```python
print(Counter(df['Nationality'].values).most_common(5))  #showing top 5 countries which send highest no of players
```

```python
bar_plot = dict(Counter(df['Nationality'].values).most_common(5)) # Bar plot for showing top-5 countries 
# zip() is a built-in Python function that gives us an iterator of tuples.
plt.bar(*zip(*bar_plot.items()))
plt.show()
```

```python
# define a function that help us plot the most common 5 values for any different attribute of data set.
def plot_most_common(category):
    bar_plot = dict(Counter(df[category].values).most_common(5))
    plt.bar(*zip(*bar_plot.items()))
    plt.show()
```

```python
plot_most_common('Position') # Function is called and displaying the Top 5 positions 
```
